```python
import pandas as pd

# 1. Load the real competition data (Path corrected!)
df = pd.read_csv('/kaggle/input/competitions/spaceship-titanic/train.csv')

# 2. Check the size of the dataset
print("Dataset Shape:", df.shape)

# 3. Find out exactly where the missing values (NaN) are hiding!
print("\n🔍 Missing Values in Live Data:")
print(df.isnull().sum())
```

```python
# 'Age' column mein jo missing values hain, unko average (median) age se bhar do
df['Age'] = df['Age'].fillna(df['Age'].median())

# 'VIP' status mein jo missing hai, usko sabse common status (False) se bhar do (Updated for new Pandas version)
df['VIP'] = df['VIP'].fillna(False).infer_objects(copy=False)

# Clean karne ke baad dobara check karo
print("\n✅ After Cleaning 'Age' and 'VIP':")
print(df[['Age', 'VIP']].isnull().sum())
```

```python
# 1. Test data (naya data jispar prediction karni hai) load karo
test_df = pd.read_csv('/kaggle/input/competitions/spaceship-titanic/test.csv')

# 2. Test data ko bhi clean karo (jaise Train data ko kiya tha)
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].median())
test_df['VIP'] = test_df['VIP'].fillna(False).infer_objects(copy=False)

# 3. Dummy Prediction banate hain (Abhi ke liye maan lete hain sab bach gaye - False)
# Asli ML model hum aage ke dino mein lagayenge!
submission = pd.DataFrame({
    'PassengerId': test_df['PassengerId'],
    'Transported': False 
})

# 4. Data ko 'submission.csv' file mein save karo (Kaggle yahi maang raha hai)
submission.to_csv('submission.csv', index=False)
print("✅ submission.csv successfully created! Ab 'Submit' button dabao!")
```
