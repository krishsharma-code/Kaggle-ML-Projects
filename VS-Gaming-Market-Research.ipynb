```python
# Import basic Data Science libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Check the file paths of the dataset we just added
print("Files in this dataset:")
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Load the dataset using the path printed above
file_path = '/kaggle/input/datasets/gregorut/videogamesales/vgsales.csv'
df = pd.read_csv(file_path)

# Display the top 5 rows of the dataset
display(df.head())
```

```python
# Graph ka size set karna
plt.figure(figsize=(12, 6))

sns.countplot(x='Genre', data=df, order=df['Genre'].value_counts().index, hue='Genre', palette='magma', legend=False)

plt.title('Most Popular Game Genres Worldwide', fontsize=16, fontweight='bold')
plt.xlabel('Game Genre', fontsize=12)
plt.ylabel('Number of Games Released', fontsize=12)
plt.xticks(rotation=45) 

# Graph screen par show karna
plt.show()
```

```python
# Top 10 Publishers by Global Sales calculate karna
top_publishers = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))

# Bar chart banana
sns.barplot(x=top_publishers.index, y=top_publishers.values, hue=top_publishers.index, palette='viridis', legend=False)

plt.title('Top 10 Game Publishers by Global Sales (in Millions)', fontsize=16, fontweight='bold')
plt.xlabel('Publisher Name', fontsize=12)
plt.ylabel('Total Global Sales', fontsize=12)
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Top 15 Platforms nikalna
sns.countplot(x='Platform', data=df, order=df['Platform'].value_counts().iloc[:15].index, hue='Platform', palette='coolwarm', legend=False)

plt.title('Top 15 Gaming Platforms by Release Count', fontsize=16, fontweight='bold')
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Number of Games', fontsize=12)

plt.show()
```
