```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 1. Smart Discovery: Automatically locate the .csv file within the directory
file_path = ""
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.endswith('.csv'):
            file_path = os.path.join(dirname, filename)
            break

print("✅ File successfully located at:", file_path)

# 2. Load the dataset
df = pd.read_csv(file_path)

# 3. Create a bar chart showing the Top 5 Gaming Genres
df['Genre'].value_counts().head(5).plot(kind='bar', color='purple', title='Top 5 Gaming Genres Worldwide')
plt.ylabel('Number of Games Sold')
plt.show()
```
