```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries Imported Successfully! ✅")

# Creating our own dictionary data for VS Gaming Studio
gaming_data = {
    "Player_Name": ["Krish", "Rahul", "Aman", "Priya", "Neha", "Rohan"],
    "Level": [15, 8, 22, 11, 25, 5],
    "Score": [1450, 800, 2100, 1050, 2500, 400],
    "Role": ["Sniper", "Medic", "Assault", "Sniper", "Assault", "Medic"]
}

# Converting dictionary to a Pandas DataFrame (Table)
df = pd.DataFrame(gaming_data)
print("Data Ready! ✅")
```

```python
# Show the first 5 rows of the data
print("--- Top Players Data ---")
display(df.head())

# Get a quick mathematical summary of the data (average score, max level, etc.)
print("\n--- Data Summary ---")
display(df.describe())
```

```python
# Filtering players with score > 1000
pro_players = df[df['Score'] > 1000]

print("--- Pro Players (Score > 1000) ---")
display(pro_players)
```

```python
# Plotting Player Scores
plt.figure(figsize=(8, 5))
plt.bar(df['Player_Name'], df['Score'], color='cyan', edgecolor='black')

# Adding Titles and Labels
plt.title("VS Gaming Studio - Player Scores", fontsize=16, color='white')
plt.xlabel("Player Names", fontsize=12)
plt.ylabel("Scores", fontsize=12)

# Setting a dark background theme for the plot (Cyberpunk vibe)
plt.style.use('dark_background')

# Show the graph
plt.show()
```
